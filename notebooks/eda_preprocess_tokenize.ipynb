```python
df = df[['Ticket Type', 'Ticket Subject', 'Ticket Description', 'Ticket Channel']]
df
```

```python
df.loc[:,'text'] = (
    'Subject: ' + df['Ticket Subject'] + ' | ' +
    'Channel: ' + df['Ticket Channel'] + ' | ' + 
    'Description: ' + df['Ticket Description']
)
df
```

```python
drop_cols = ['Ticket Subject', 'Ticket Description', 'Ticket Channel']
df = df.drop(drop_cols, axis=1)
df
```

```python
import re
def basic_clean(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)
    text = re.sub(r"<.*?>", "", text)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df['clean_text'] = df['text'].apply(basic_clean)
df
```

```python
label2id = {
    'Technical issue': 0,
    'Billing inquiry': 1, 
    'Cancellation request': 2,
    'Product inquiry': 3,
    'Refund request': 4
}

df['label'] = df['Ticket Type'].map(label2id)
df
```
